```python
import numpy as np
import matplotlib.pyplot as plt
```

Para generar una se√±al cuya amplitud est√© modulada:

$$c(t) = A_c cos (ùúî_ct)$$ se√±al original (portadora)

$$m(t) = A_m cos(ùúî_m t) $$ se√±al moduladora

$$AM(t) = [A_c + m(t)] cos(ùúî_c t) $$ Se√±al de Amplitud Modulada

# Onda Resultante de las Componentes Armonicas de Marea 
  usando las amplitudes derivadas de la teoria de equilibrio 

(a) periodo de 30 dias, onda de 1m de amplitud
con frecuencia de 28.9841 ¬∞/h (per√≠odo de 12.42 h).

```python
t=np.arange(0,24*30) # Eje temporal de un mes de datos horarios
M2=1*np.cos(np.deg2rad(28.9841*t))
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot(t/24,M2, color= "m")
axs.set(title='M2',ylabel='Altura (m)',xlabel='d√≠as')
axs.grid(which='both',ls=':')
```

(b)  Para el mismo perioo una onda de 0.4652 m de amplitud con una frecuencia de 30¬∞/h (per√≠odo de 12 h).

```python
S2=0.4652*np.cos(np.deg2rad(30*t))
suma=S2+M2
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot(t/24,M2,color="m")
axs.plot(t/24,suma, color="coral")
axs.set(title='M2+S2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(['M2','M2 + S2'])
axs.grid(which='both',ls=':')
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot(t/24,S2, color= "darkcyan")
axs.set(title='S2',ylabel='Altura (m)',xlabel='d√≠as')
axs.grid(which='both',ls=':')
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot(t/24,M2,color="m")
axs.plot(t/24,S2, color="darkcyan")
axs.set(title='M2 y S2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(['M2',' S2'])
axs.grid(which='both',ls=':')
```

(c) para la onda suma obtenida, le sumo otra  onda de 0.1915 m de amplitud y per√≠odo de 12.66 h.

```python
N2=0.195*np.cos(np.deg2rad(28.436001*t))
suma_1= suma + N2
```

(d) a la onda resultante en c se le agregan dos ondas de per√≠odos 23.93 h y 25.82 h, ambas con amplitudes de 0.40 y con fases 90¬∞ y 0¬∞, respectivamente. 

```python
K1=0.40*np.cos(np.deg2rad(15.043877*t + 90))
O1=0.40*np.cos(np.deg2rad(13.94268*t))
suma_2= suma_1 + K1 + O1
```

```python
t=np.arange(0,24*365) # Eje temporal de un mes de datos horarios
S2=0.47*np.cos(np.deg2rad(30*t))
T2= 0.0587*np.cos(np.deg2rad(29.9589*t))
R2=0.0084*np.cos(np.deg2rad(30.0411*t + 180))
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot((t/24),S2,color="r")
axs.plot((t/24),T2, color="c")
axs.plot((t/24), R2, color="g")
axs.set(title='S2  y T2 y R2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(['S2','T2', "R2"])
axs.grid(which='both',ls=':')
```

```python
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot((t/24)[0:100],S2[0:100],color="r")
axs.plot((t/24)[0:100],T2[0:100], color="c")
axs.plot((t/24)[0:100], R2[0:100], color="g")
axs.set(title='S2  y T2 y R2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(['S2','T2', "R2"])
axs.grid(which='both',ls=':')
```

```python
Suma_4= S2 + T2 + R2
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot((t/24)[0:100],S2[0:100],color="m")
axs.plot((t/24)[0:100],Suma_4[0:100], color="coral")
axs.set(title='S2 +T2+R2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(["S2",'S2 + T2 +R2'])
axs.grid(which='both',ls=':')
```

```python
Suma_4= S2 + T2 + R2
fig,axs=plt.subplots(1,1,figsize=(12,4))
axs.plot((t/24),S2,color="m", alpha=0.5)
axs.plot((t/24),Suma_4, color="coral", alpha=0.5)
axs.set(title='S2 +T2+R2',ylabel='Altura (m)',xlabel='d√≠as')
axs.legend(["S2",'S2 + T2 +R2'])
axs.grid(which='both',ls=':')
```
