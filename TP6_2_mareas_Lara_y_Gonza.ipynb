```python
mas=residuos.loc[(residuos.residuos>=umbral) | (residuos.residuos<=-umbral)]
#menos=residuos.loc[residuos.residuos<=-umbral]
```

```python
tabla=pd.DataFrame(columns=['inicio','final','pico','momento'])
```

```python
nom2[i]
```

```python
if nom2[i]=='Gonzalo':
  mas['dT'] = mas.index.to_series().diff().astype('timedelta64[s]')
else:
  mas['dT'] = mas.index.to_series().diff().astype('timedelta64[h]')
```

```python
tabla['inicio']=mas.loc[mas.dT!=delta].index
```

```python
fin=np.where(mas.dT!=1)[0]-1
tabla['final']=mas.iloc[np.roll(fin,-1)].index
```

```python
datos=datos.set_index(datos.FECHA,drop=True)
```

```python
tabla['momentonivel']=np.nan
for e in range(len(tabla)):
  tabla.momento.iloc[e]=mas.residuos.loc[tabla.inicio.iloc[e]:tabla.final.iloc[e]].abs().idxmax()
  tabla.pico.iloc[e]=mas.residuos.loc[tabla.momento.iloc[e]]
  tabla.momentonivel.iloc[e]=datos.OBS.loc[tabla.inicio.iloc[e]:tabla.final.iloc[e]].abs().max()
  
```

Transformada de fourier

```python
residuos.info()
```

```python
Y_residuo=np.fft.fft(residuos.values)
```

```python
xff_residuo=(1/len(Y_residuo))*np.abs(Y_residuo)#Y_ushu.real
```

```python
freq_residuo.size
```

```python
xff_residuo.size
```

```python
fig, axs=plt.subplots(1,1,figsize=(12,4))
axs.plot(freq_residuo,xff_residuo*2,':o')
axs.set(ylabel='Amplitud (m)',xlabel='frecuencias (Hz)')
axs.grid(color='grey',ls=':',which='both')
```
